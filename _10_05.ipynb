```python
import random
n=4
matris=[]
for i in range(n):
    satir=[]
    for j in range(n):
        satir.append(random.randint(1,10))
        matris.append(satir)
    print(matris)

        
```

```python

for i in range(n):
    for j in range(n):
        if i<j:
            matris[i][j]=0
```

```python

for i in range(n):
    for j in range(n):
        print(f"{matris[i][j]:4d}",end=' ')
    print("")
```
