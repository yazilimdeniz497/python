```python
import random
gizli_sayi=random.randint(0,20)
deneme_sayisi=0
while True:
    tahmin=int(input("tahmini sayi giriniz: "))
    deneme_sayisi+=1
    if tahmin<gizli_sayi:
        print("lütfen daha büyük tahmin sayisi giriniz: ")
    elif tahmin>gizli_sayi:
        print("lütfen daha küçük tahmin sayisi giriniz: ")
    else:
        print(f"Tebrikler! {deneme_sayisi} adımda tahmin ettiniz")
        break
    
        
```
