```python
toplam=0
print("Pozitif sayilar giriniz (Bitirmek için 0 girin)")
while True:
     girilen_sayi=int(input("Lütfen pozitif sayı giriniz: "))
     if girilen_sayi==0:
         break
     elif girilen_sayi>0:
         toplam+=girilen_sayi
     else:
         print(f"negatif sayı yok sayıldı")
     print(f"girilen sayilarin toplami: {toplam}")
```
